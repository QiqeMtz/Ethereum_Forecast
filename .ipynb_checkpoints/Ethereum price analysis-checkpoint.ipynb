```python
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

#from pandas import datetime
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error

plt.rcParams["figure.figsize"] = "21, 10"
plt.rcParams.update({'font.size': 15})
```

```python
#dateparse = lambda dates: pd.to_datetime(dates, format='%Y-%m-%d')
def dateparse (time_in_secs):    
    return dt.datetime.fromtimestamp(float(time_in_secs))

eth_series = pd.read_csv('etherprice.csv',parse_dates=True, date_parser=dateparse, index_col='Date', names=['Date', 'Price'], squeeze=True)

eth_series.index = eth_series.index.strftime('%Y-%m-%d')
eth_series.index = eth_series.index.astype('datetime64[ns]')
eth_series
```

```python
etherneum_plot_original = plt.plot(eth_series, color='blue')
#plt.show()

eth_original = mpatches.Patch(color='blue', label='Original Ethereum price')

plt.legend(handles=[eth_original])

plt.xlabel('Time (daily)')
plt.ylabel('Price (USD)')
plt.title('Ethereum daily price')

plt.margins(0.005, 0)
#plt.pyplot.savefig('ARI_diagnoses_weekly.eps', format='eps', dpi=1000)
plt.show()
```

```python
# Usado para entrenar el modelo
# X tiene todos los valores Fecha -  Numero de Casos (364 en total) de inicio a fin
X = eth_series

# size almacena el valor 240, equivale al 66.66% de todos los valores (66.6% de 364) 
size = int(len(X) * 0.66)

# train [fecha, casos] tiene desde el valor 101 al 411 (240 valores) que van desde el 2009-04-26 al 2013-10-25
# y va desde 0 a size (240)
# test almacena los datos [fecha, casos] desde el valor 344 al 268 (124 valores) que van desde el 2013-11-01 al 2016-03-28
# y va desde size(240) a len(X) (364)
train, test = X[0:size], X[size:len(X)]

# history almacena la lista de 364 valores (todos) [101-268], sin considerar la fecha, solo una lista de numeros (casos)
history = [x for x in train]

# predictions es una lista donde se hara un append de los valores predichos
ARIMA_predictions = list()
ARIMA_predictions_date = list()
```

```python
for t in range(len(test)):
	ARIMA_model = ARIMA(history, order=(0,1,2))
	ARIMA_model_fit = ARIMA_model.fit(disp=0)
	output = ARIMA_model_fit.forecast()
	yhat = output[0]
	ARIMA_predictions.append(yhat)
	obs = test[t]
	dates = test.index[t]
	ARIMA_predictions_date.append(dates)
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
#error = mean_squared_error(test, ARIMA_predictions)
#real_error = np.sqrt(error)
#print('Test MSE: %.3f' % real_error)
```

```python
# Aqui se necesita convertir a serie de tiempo, para lograr tener fechas y valores predecidos de ARIMA
ARIMA = pd.DataFrame({'Date':ARIMA_predictions_date, 'Predicted':ARIMA_predictions, 'Expected': test})
```

```python
#Ploteo
ETH_expected, = plt.plot(ARIMA['Expected'], label="Original Ethereum price", color="red")
ETH_predicted, = plt.plot(ARIMA['Predicted'], label="Ethereum Forecasted Price", color='blue', linestyle='--')

original_legend = mpatches.Patch(color='red', label='Original Ethereum price')
ARIMA_legend = mpatches.Patch(color='blue', label='Forecasted Ethereum price')
plt.legend(handles=[original_legend, ARIMA_legend])

plt.xlabel('Time (Days)')
plt.ylabel('Price (USD)')
plt.title('Ethereum forecasting')

plt.margins(0.005, 0)
#plt.savefig('ARIMA_VS_ORIGINAL.eps', format='eps', dpi=1000)
plt.show()
```

```python
# Para predecir
# X tiene todos los valores Fecha -  Numero de Casos (364 en total) de inicio a fin
Xp = eth_series

# size almacena el valor 240, equivale al 66.66% de todos los valores (66.6% de 364) 
#sizep = int(len(Xp) * 0.66)

# train [fecha, casos] tiene desde el valor 101 al 411 (240 valores) que van desde el 2009-04-26 al 2013-10-25
# y va desde 0 a size (240)
# test almacena los datos [fecha, casos] desde el valor 344 al 268 (124 valores) que van desde el 2013-11-01 al 2016-03-28
# y va desde size(240) a len(X) (364)
trainp = Xp

# history almacena la lista de 364 valores (todos) [101-268], sin considerar la fecha, solo una lista de numeros (casos)
historyp = [x for x in trainp]

# predictions es una lista donde se hara un append de los valores predichos
ARIMA_predictionsp = list()
ARIMA_predictions_datep = list()
```

```python
new_dates = pd.date_range('20170720','20171231')
new_dates.format(formatter=lambda x: x.strftime('%Y-%m-%d'))
new_dates
```

```python
for t in range(len(new_dates)):
	ARIMA_model = ARIMA(historyp, order=(0,1,2))
	ARIMA_model_fit = ARIMA_model.fit(disp=0)
	output = ARIMA_model_fit.forecast()
	yhat = output[0]
	ARIMA_predictionsp.append(yhat)
	#obs = test[t]
	historyp.append(yhat)
	print('predicted=%f' % (yhat))
```
